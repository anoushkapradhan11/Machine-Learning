```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("cereal.csv")
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(df['sugars'], bins=10, color='skyblue', edgecolor='black')
plt.title('Sugar Content in Cereals')
plt.xlabel('Sugar')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(df['vitamins'], bins=10, color='orange', edgecolor='black')
plt.title('Vitamin Content in Cereals')
plt.xlabel('Vitamins')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
manufacturer_map = {
    'N': 'Nabisco',
    'Q': 'Quaker Oats',
    'K': 'Kelloggs',
    'R': 'Raslston Purina',
    'G': 'General Mills',
    'P': 'Post',
    'A': 'American Home Foods Products'
}
df['manufacturer_full'] = df['mfr'].map(manufacturer_map)
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.countplot(y='manufacturer_full', data=df, palette='Set2', order=df['manufacturer_full'].value_counts().index)
plt.title("Number of Cereals by Manufacturer")
plt.xlabel("Count")
plt.ylabel("Manufacturer")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x = df.select_dtypes(include='number').drop(columns=['rating'])
y = df['rating']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
mse = mean_squared_error(y_test, y_pred)

print(f"Mean Squared Error on test data: {mse:.2f}")
```
