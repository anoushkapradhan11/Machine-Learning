```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
df = pd.read_csv("FyntraCustomerData.csv")
```

```python
sns.jointplot(data=df, x='Time_on_Website', y='Yearly_Amount_Spent', kind='reg', color='blue')
plt.suptitle("Time on Website vs Yearly Amount Spent", y=1.02)
plt.show()
```

```python
sns.jointplot(data=df, x='Time_on_App', y='Yearly_Amount_Spent', kind='reg', color='green')
plt.suptitle("Time on App vs Yearly Amount Spent", y=1.02)
plt.show()
```

```python
sns.pairplot(df[['Avg_Session_Length', 'Time_on_App', 'Time_on_Website',
                 'Length_of_Membership', 'Yearly_Amount_Spent']])
plt.suptitle("Pairplot of All Numerical Features", y=1.02)
plt.show()
```

```python
sns.lmplot(data=df, x='Length_of_Membership', y='Yearly_Amount_Spent', line_kws={'color': 'red'})
plt.title("Length of Membership vs Yearly Amount Spent")
plt.show()
```

```python
X = df[['Avg_Session_Length', 'Time_on_App', 'Time_on_Website', 'Length_of_Membership']]
y = df['Yearly_Amount_Spent']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=85)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

plt.figure(figsize=(8,6))
plt.scatter(y_test, y_pred, edgecolors='black', color='skyblue')
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--')
plt.xlabel("Actual Yearly Amount Spent")
plt.ylabel("Predicted Yearly Amount Spent")
plt.title("Actual vs Predicted Spending")
plt.show()
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
```
